```python
videogame_path = "Resources/VideoGamesSales.csv"

vg_df = pd.read_csv(videogame_path, encoding ='iso-8859-1')
vg_df.head()
```

```python
vg_df
```

```python
#Remove unecessary columns and save in a new DataFrame
vgcolremove = vg_df.drop(columns=["Critic_Score","Critic_Count", "User_Score", "User_Count", "Developer", "Rating"])
vgcolremove
```

```python
vgcolrename = vgcolremove.rename(columns = {"Name": "Game",
                                            "Platform": "Console Name",
                                             "Year_of_Release":"Year",
                                             "Genre": "Genre",
                                             "Publisher": "Publisher",
                                             "NA_Sales":"North America",
                                             "EU_Sales":"Europe",
                                             "JP_Sales": "Japan",
                                             "Other_Sales":"Rest of World",
                                             "Global_Sales":"Global",                                         
                                           
                                           })
vgcolrename
```

```python
vgnona = vgcolrename.dropna()
vgnona
```

## VIDEO GAMES CLEAN TABLE

```python
vgnodups = vgnona.drop_duplicates(subset=["Game", "Year"])
vgnodups
```

```python
vgnodupsy = vgnodups.sort_values(by="Year", ascending=False)
vgnodupsy
```

```python
vgnodupsy.dtypes
```
