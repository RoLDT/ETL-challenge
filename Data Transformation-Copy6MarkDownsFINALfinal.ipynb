## Database Structure: Star Schema

## Games Catalogue

```python
games_df = videogames_df["Game"].values
games_df
```

```python
games_df = videogames_df["Game"].unique()
games_df
```

```python
gamesdim_df = pd.DataFrame(games_df)
gamesdim_df
```

```python
gamesdim_df.index = [x for x in range(1, len(gamesdim_df.values)+1)]
#gamesdim_df.rename(columns={0:"game_name"}, inplace = True),
```

```python
gamesdim_df = gamesdim_df.reset_index()
gamesdim_df
```

```python
gamesdim_df.rename(columns={"index":"game_id",0:"game_name"}, inplace = True)
gamesdim_df
```

## Catalogue Console

```python
console_df= videogames_df["Console Name"].unique()
console_df
```

```python
consoledim_df = pd.DataFrame(console_df)
consoledim_df
```

```python
consoledim_df.index = [x for x in range(1, len(consoledim_df.values)+1)]
```

```python
consoledim_df = consoledim_df.reset_index()
consoledim_df
```

```python
consoledim_df.rename(columns={"index":"console_id",0:"console_name"}, inplace = True)
consoledim_df
```

## Publisher Catalogue

```python
publisher_df = videogames_df["Publisher"].unique()
publisher_df
```

```python
publisherdim_df = pd.DataFrame(publisher_df)
publisherdim_df
```

```python
publisherdim_df.index = [x for x in range(1, len(publisherdim_df.values)+1)]
```

```python
publisherdim_df = publisherdim_df.reset_index()
publisherdim_df
```

```python
publisherdim_df.rename(columns={"index":"publisher_id",0:"publisher"}, inplace = True)
publisherdim_df
```

```python
publisherdim_df
```

## Genre Catalogue

```python
genre_df = videogames_df["Genre"].unique()
genre_df
```

```python
genredim_df = pd.DataFrame(genre_df)
genredim_df
```

```python
genredim_df.index = [x for x in range(1, len(genredim_df.values)+1)]
```

```python
genredim_df = genredim_df.reset_index()
genredim_df
```

```python
genredim_df.rename(columns={"index":"genre_id",0:"genre"}, inplace = True)
genredim_df
```

## Facts Table

```python
videogames_merged = pd.merge(videogames_df, gamesdim_df, left_on ='Game',  right_on='game_name' , how ='left')
videogames_merged2 = pd.merge(videogames_merged, consoledim_df, left_on ='Console Name',  right_on='console_name' , how ='left')
videogames_merged3 = pd.merge(videogames_merged2, publisherdim_df, left_on ='Publisher',  right_on='publisher' , how ='left')
videogames_merged4 = pd.merge(videogames_merged3, genredim_df, left_on ='Genre',  right_on='genre' , how ='left')
```

```python
videogames_merged4.head()
```

```python
videogames_final = videogames_merged4[['game_id','genre_id','publisher_id','console_id', 'Year', 'North America','Europe','Japan','Rest of World']]
```

```python
videogames_final
```

```python
videogames_final = videogames_final.rename(columns={"Year":"year", "North America":"na_sales", "Europe": "eu_sales", "Japan": "jp_sales", "Rest of World": "other_sales"})
```

```python
videogames_final
```

```python
videogames_final = videogames_final.reset_index()
videogames_final
```

```python
videogames_final = videogames_final.rename(columns={"index":"sales_id"})
videogames_final
```
